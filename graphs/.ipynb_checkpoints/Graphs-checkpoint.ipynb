```python
import matplotlib.pyplot as plt

# Data
threads = [10, 10, 10, 16, 16]
vector_sizes = [50, 100, 500, 50, 100]
execution_times = [0.001, 0.001, 0.001, 0.001, 0.001]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data
ax.plot(threads, execution_times, marker='o', linestyle='-')

# Set axis labels and title
ax.set_xlabel('Number of Threads')
ax.set_ylabel('Execution Time (seconds)')
ax.set_title('Execution Time vs. Number of Threads for Different Vector Sizes')

# Customize the x-axis tick labels
ax.set_xticks(threads)
ax.set_xticklabels(threads)

# Add a legend
ax.legend(['Vector Size 50', 'Vector Size 100', 'Vector Size 500'])

# Show the plot
plt.grid(True)
plt.show()
```

```python
samples = [10, 50, 100, 10, 50, 100]
threads = [5, 5, 5, 10, 10, 10]
pi_estimates = [4, 3.6, 3.4, 4, 3.2, 3.6]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data with different threads as separate lines
for thread_count in set(threads):
    x = [samples[i] for i in range(len(samples)) if threads[i] == thread_count]
    y = [pi_estimates[i] for i in range(len(pi_estimates)) if threads[i] == thread_count]
    ax.plot(x, y, marker='o', linestyle='-', label=f'{thread_count} Threads')

# Set axis labels and title
ax.set_xlabel('Number of Samples')
ax.set_ylabel('Estimated Value of Pi')
ax.set_title('Estimated Value of Pi vs. Number of Samples for Different Thread Counts')

# Add a legend
ax.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Data
vector_sizes = [100, 500, 1000, 1000, 100, 500, 500, 1000, 10000, 100000]
threads = [8, 8, 8, 8, 16, 16, 16, 16, 16, 16]
time_taken = [0.001, 0.002, 0.000, 0.001, 0.001, 0.000, 0.001, 0.001, 0.001, 0.001]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data with different threads as separate lines
for thread_count in set(threads):
    x = [vector_sizes[i] for i in range(len(vector_sizes)) if threads[i] == thread_count]
    y = [time_taken[i] for i in range(len(time_taken)) if threads[i] == thread_count]
    ax.plot(x, y, marker='o', linestyle='-', label=f'{thread_count} Threads')

# Set axis labels and title
ax.set_xlabel('Number of Elements in Vectors')
ax.set_ylabel('Time Taken (seconds)')
ax.set_title('Time Taken vs. Number of Elements in Vectors for Different Thread Counts')

# Add a legend
ax.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Data
data = {
    'Schedule': ['STATIC', 'STATIC', 'STATIC', 'STATIC', 'STATIC', 'STATIC', 'STATIC', 'STATIC', 'STATIC', 'STATIC', 'DYNAMIC', 'DYNAMIC', 'DYNAMIC', 'DYNAMIC', 'DYNAMIC', 'DYNAMIC', 'DYNAMIC', 'DYNAMIC', 'DYNAMIC', 'DYNAMIC'],
    'Chunk Size': [1, 11, 21, 31, 41, 51, 61, 71, 81, 91, 1, 11, 21, 31, 41, 51, 61, 71, 81, 91],
    'Number of Threads': [8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8],
    'Execution Time (seconds)': [0, 0, 0, 0, 0.002, 0, 0, 0.001, 0, 0, 0, 0, 0, 0, 0, 0.001, 0, 0, 0, 0],
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data for each schedule
schedules = df['Schedule'].unique()
for schedule in schedules:
    df_schedule = df[df['Schedule'] == schedule]
    ax.plot(df_schedule['Chunk Size'], df_schedule['Execution Time (seconds)'], marker='o', linestyle='-', label=f'{schedule}')

# Set axis labels and title
ax.set_xlabel('Chunk Size')
ax.set_ylabel('Execution Time (seconds)')
ax.set_title('Execution Time vs. Chunk Size for Different Schedules (8 Threads)')

# Add a legend
ax.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python

data = {
    'Fibonacci Number': [20, 25, 30, 20, 25, 30],
    'Number of Threads': [8, 8, 8, 16, 16, 16],
    'Execution Time (seconds)': [0.011, 0.093, 1.095, 0.014, 0.119, 1.374],
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data for each number of threads
threads = df['Number of Threads'].unique()
for thread in threads:
    df_thread = df[df['Number of Threads'] == thread]
    ax.plot(df_thread['Fibonacci Number'], df_thread['Execution Time (seconds)'], marker='o', linestyle='-', label=f'{thread} Threads')

# Set axis labels and title
ax.set_xlabel('Fibonacci Number')
ax.set_ylabel('Execution Time (seconds)')
ax.set_title('Execution Time vs. Fibonacci Number for Different Thread Counts')

# Add a legend
ax.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
data = {
    'Producers': [10, 20, 5],
    'Consumers': [10, 20, 5],
    'Execution Time (seconds)': [2.08, 3.169, 1.522],
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
for index, row in df.iterrows():
    label = f'{row["Producers"]} Producers, {row["Consumers"]} Consumers'
    ax.bar(label, row['Execution Time (seconds)'])

# Set axis labels and title
ax.set_xlabel('Producers and Consumers')
ax.set_ylabel('Execution Time (seconds)')
ax.set_title('Execution Time vs. Number of Producers and Consumers')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
